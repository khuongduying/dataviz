### Creating a ggplot

Create an empty graph, since we haven't told it how to visualize yet, for now it's empty.

```r
ggplot(data = penguins)
```

Now we will only map flipper length to the x aesthetic and body mass to the y aesthtic.

```r
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
)
```

Add a layer of points, create a scatterplot.

```r
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
) +
  geom_point()
```

### Adding aesthetics and layers

Represent species with different colored points.
Does the relationship between flipper length and body mass differ by species?

```r
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g, color = species)
) +
  geom_point()
```

Add one more layer - a smooth curve displaying the relationship between body mass and flipper length.

```r
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g, color = species)
) +
  geom_point() +
  geom_smooth(method = "lm") ## line of best fit based on a linear model
```

We want points to be colored based on species but don't want the lines to be separated out for them, we should specify color = species for geom_point() only.

```r
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
) +
  geom_point(mapping = aes(color = species)) +
  geom_smooth(method = "lm")
```

Use different shapes for each species of penguins and improve labels

```r
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
) +
  geom_point(mapping = aes(color = species, shape = species)) +
  geom_smooth(method = "lm")
```

Improve the labels of the plot using the labs() function in a new layer.

Improve the color palette to be colorblind safe with the scale_color_colorblind() function from the ggthemes package.


```r
ggplot(
  data = penguins,
  mapping = aes(x = flipper_length_mm, y = body_mass_g)
) +
  geom_point(aes(color=species, shape=species)) +
  geom_smooth(method="lm") +
  labs(
    title = "Body mass and flipper length",
    subtitle = "Dimensions for Adelie, Chinstrap, and Gentoo Penguins",
    x = "Flipper length (mm)", y = "Body mass (g)",
    color = "Species", shape = "Species"
  ) +
  scale_color_colorblind()
```
